# you can also pass another two parameters that defines Row Indexes and column names

```python
df1 = pandas.DataFrame([[2,4,6],[4,16,36]],["Number","Square"],["Num 1","Num 2","Num 3"])
df1
```

# You can define Columns and Row Indexes with 'columns' & 'index' parameters

```python
df2 = pandas.DataFrame([[2,4,6],[4,16,36]],columns=["Num 1","Num 2","Num 3"],index=["Number","Square"])
df2
```

# You Can also pass the "Disctionary" to DataFrame
